```python
from sklearn import metrics

def calculate_accuracy(classifier, x_train, x_test, y_train, y_test):  # Accept y_train and y_test separately
    classifier.fit(x_train, y_train)  # Fit the model
    y_pred_train = classifier.predict(x_train)  # Predict for train set
    y_pred_test = classifier.predict(x_test)  # Predict for test set
    
    # Calculate accuracies
    accuracy_train = metrics.accuracy_score(y_train, y_pred_train)
    accuracy_test = metrics.accuracy_score(y_test, y_pred_test)
    
    # Print the accuracies
    print('Train accuracy:', '{:.2f}'.format(accuracy_train))
    print('Test accuracy:', '{:.2f}'.format(accuracy_test))    
    # Return results
    return accuracy_train, accuracy_test, classifier 

# Example usage
# Ensure 'classifier', 'x_train', 'x_test', 'y_train', and 'y_test' are defined
accuracy_train, accuracy_test, trained_classifier = calculate_accuracy(classifier, x_train, x_test, y_train, y_test)
```

```python
x= x.reset_index(drop=True)
y = y.reset_index(drop=True)
```

```python
print("X shape:", x.shape)
print("y shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
print("X_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", x_test.shape)
print("y_test shape:", y_test.shape)
```

```python
#cree et entainer et visualiser un arbre de décision
classifier = DecisionTreeClassifier(max_depth=2, random_state=42, criterion='entropy')
accuracy_train, accuracy_test, trained_classifier = calculate_accuracy(classifier, x_train, x_test, y_train, y_test)

print("Accuracy Train:", accuracy_train)
print("Accuracy Test:", accuracy_test)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

# Créer un modèle d'arbre de décision
classifier = DecisionTreeClassifier(max_depth=2, random_state=42, criterion='entropy')

# Entraîner le modèle
classifier.fit(x_train, y_train)

# Prédire sur les données d'entraînement et de test
y_train_pred = classifier.predict(x_train)
y_test_pred = classifier.predict(x_test)

# Calculer la précision
accuracy_train = accuracy_score(y_train, y_train_pred)
accuracy_test = accuracy_score(y_test, y_test_pred)

print("Accuracy Train:", accuracy_train)
print("Accuracy Test:", accuracy_test)

# Visualiser l'arbre
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plot_tree(classifier, feature_names=x.columns, class_names=y.unique(), precision=2, filled=True)
plt.show()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

# Afficher la matrice de confusion
ConfusionMatrixDisplay.from_estimator(classifier, x_test, y_test)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Créer le modèle de Random Forest
rf_classifier = RandomForestClassifier(max_depth=2, n_estimators=20, random_state=42, criterion='entropy')

# Entraîner le modèle
rf_classifier.fit(x_train, y_train)

# Calculer les métriques
y_test_pred_rf = rf_classifier.predict(x_test)
accuracy_train_rf = rf_classifier.score(x_train, y_train)
accuracy_test_rf = rf_classifier.score(x_test, y_test)

print("Random Forest - Accuracy Train:", accuracy_train_rf)
print("Random Forest - Accuracy Test:", accuracy_test_rf)

# Afficher la matrice de confusion pour Random Forest
ConfusionMatrixDisplay.from_estimator(rf_classifier, x_test, y_test)
plt.show()
```

```python
from sklearn.metrics import RocCurveDisplay

# Afficher la courbe ROC
RocCurveDisplay.from_estimator(rf_classifier, x_test, y_test)
plt.show()
```

```python
from sklearn.metrics import classification_report

# Afficher le rapport
report = classification_report(y_true=y_test, y_pred=y_test_pred_rf, target_names=y.unique())
print(report)
```
